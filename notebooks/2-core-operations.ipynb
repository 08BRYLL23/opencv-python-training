```python
import cv2
image = cv2.imread('../assets/Test.jpg',cv2.IMREAD_COLOR)
howie = cv2.imread('../assets/howie.png',cv2.IMREAD_COLOR)
bernie = cv2.imread('../assets/Test.png',cv2.IMREAD_COLOR)
big_bang = cv2.imread('../assets/Trevor.jpg',cv2.IMREAD_COLOR)

#whitespace removing

def crop_image():
    from_center = False
    ROI = cv2.selectROI(image,from_center)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    crops = image[ROI[1]:(ROI[1]+ROI[3]),ROI[0]:(ROI[0]+ROI[2])]
    cv2.namedWindow('Before',cv2.WINDOW_NORMAL)
    cv2.imshow('Before',crops)
    cv2.imwrite('Trevor.jpg',crops)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
def show_image(image):
    cv2.namedWindow('Filtered',cv2.WINDOW_NORMAL)
    cv2.imshow('Filtered',image)
    cv2.imwrite('Delfin-Bryll.png',image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


rows,cols,shape = big_bang.shape
roi = bernie[0:rows,0:cols]
#creating a mask
big_bang_gray = cv2.cvtColor(big_bang, cv2.COLOR_BGR2GRAY)
ret, mask = cv2.threshold(big_bang_gray,220,255,cv2.THRESH_BINARY_INV)
# show_image(mask)
#inverting the mask
mask_inv = cv2.bitwise_not(mask)
# show_image(mask_inv)
#adding it to image and merge it
bernie_bg = cv2.bitwise_and(roi ,roi ,mask=mask_inv)
# show_image(bernig_bg)

bigbang_fg = cv2.bitwise_and(big_bang, big_bang, mask=mask)

dst = cv2.add(bernie_bg,bigbang_fg)
bernie[0:rows,0:cols]=dst
show_image(bernie)

    
# graypic_test = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
# show_image(blended)
```

```python
#logo with video
cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    bernie = frame
    rows,cols,shape = big_bang.shape
    roi = bernie[0:rows,0:cols]
    #creating a mask
    big_bang_gray = cv2.cvtColor(big_bang, cv2.COLOR_BGR2GRAY)
    ret, mask = cv2.threshold(big_bang_gray,220,255,cv2.THRESH_BINARY_INV)
    # show_image(mask)
    #inverting the mask
    mask_inv = cv2.bitwise_not(mask)
    # show_image(mask_inv)
    #adding it to image and merge it
    bernie_bg = cv2.bitwise_and(roi ,roi ,mask=mask_inv)
    # show_image(bernig_bg)
    bigbang_fg = cv2.bitwise_and(big_bang, big_bang, mask=mask)
    dst = cv2.add(bernie_bg,bigbang_fg)
    bernie[0:rows,0:cols]=dst
    cv2.imshow('VIDEO',bernie)
    if cv2.waitKey(1) == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
